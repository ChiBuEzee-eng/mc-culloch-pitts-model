```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
```

```python
#creating a data frame. The Logic here is to deny entry to those  putting on either a cap or watch but to grant access to any putting on sunglass or shoes

data = {'Names': ['Ada', 'Obi', 'John', 'Chukwuma', 'David', 'Mary', 'Collins', 'Chibueze', 'Joshua', 'Kelvin','Edward','Felix','James','Emeka', 'Mark', 'Great', 'Clinton', 'Esther', 'Deborah', 'Amaka', 'Nkiru'], 'Cap?': ['Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No','Yes','No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No'], 'Watch?': ['Yes', 'No', 'No', 'No', 'No', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'No','No', 'No', 'No', 'No', 'Yes', 'No', 'Yes'], 'Sunglass?': ['No', 'Yes', 'No', 'No', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'No', 'No', 'No', 'No', 'No', 'No', 'No','Yes', 'Yes', 'No', 'No'], 'Shoes?' : ['No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'No' ], 'Entry?': ['No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No','No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No']}
df = pd.DataFrame(data)
df
```

```python
#Encoding the Values of Yes and No to numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Cap?'] = le.fit_transform(df['Cap?'])
df['Watch?'] = le.fit_transform(df['Watch?'])
df['Sunglass?'] = le.fit_transform(df['Sunglass?'])
df['Shoes?'] = le.fit_transform(df['Shoes?'])
df['Entry?'] = le.fit_transform(df['Entry?'])
df
```

```python
#From close observation, Yes = 1, No = 0
#Changing the data frame to float
df['Cap?'] = df['Cap?'].astype("float")
df['Watch?'] = df['Watch?'].astype("float")
df['Sunglass?'] = df['Sunglass?'].astype("float")
df['Shoes?'] = df['Shoes?'].astype("float")
df['Entry?'] = df['Entry?'].astype("float")
df
```

```python
#Splitting into Independent and dependent Variable
X = df.iloc[:, [1, 2, 3, 4]].values
y = df.iloc[:, -1].values
```

```python

#Defining mcculloh_pitts_neuron
def mcculloh_pitts_neuron(inputs, weights, threshold):
    weighted_sum = tf.reduce_sum(inputs * weights)
    if weighted_sum >= threshold:
        return 1.0
    else:
        return 0.0

#initialise weights and threshold
weights = tf.constant([1, 1, 1, 1], dtype=tf.float32)
threshold = tf.constant(0.5, dtype=tf.float32)

#Training process: Adjust weights manually for simplicity
learning_rate = 0.1
epochs = 10

for epoch in range(epochs):
    print(f"Epoch {epoch + 1}")
    for inputs, expected_output in zip(X, y):

    #Get the current prediction from the neuron
        prediction = mcculloh_pitts_neuron(inputs, weights, threshold)
    #Calculate the error
        error = expected_output - prediction
    #Update the weights
        weights += learning_rate * error * inputs
        print(f"inputs:{inputs}, prediction:{prediction}, Error:{error}, Updated_Weights:{weights.numpy()}")
print("\n Training Completed\n")
```

```python
#Looking at our Model, The Machine has successfully learnt except where all the inputs were No
```
